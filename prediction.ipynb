```python
import tensorflow as tf
import os, cv2,numpy as np
from deeplab import DeepLabV3Plus
import matplotlib.pyplot as plt
```

```python
num_classes = 23
H = 512
W = 512
```

```python
model = DeepLabV3Plus(H, W, num_classes)
```

```python
model.load_weights('../../DeepLabV3_Plus-Tensorflow2.0/checkpoints/weights_drone.h5')
```

```python
img = cv2.imread('/scratch/aniketjoshi/semantic_drone_dataset/original_images/015.jpg')
image = tf.cast(img, tf.uint8)
image = tf.image.resize(image,(H, W))
image = tf.cast(image, tf.keras.backend.floatx())
image.set_shape([H,W,3])
normalized_image = image-[103.939, 116.779, 123.68]
normalized_image = np.array(normalized_image)
normalized_image = normalized_image.reshape((1, H, W, 3))
y = model.predict(normalized_image)
```

```python
y = y.reshape((H, W, num_classes))
```

```python
final = np.zeros((H, W))
```

```python
for i in range(H):
    for j in range(W):
        maxi = 0
        maxv = -1
        for k in range(num_classes):
            if(y[i][j][k] > maxv):
                maxi = k
                maxv = y[i][j][k]
        print(maxi)
        final[i][j] = maxi
```

```python
plt.imshow(final)
```
